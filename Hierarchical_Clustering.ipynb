# **Hierarchical Clustering**

## **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## **Importing DataSet**

```python
dataset = pd.read_csv('Amazon_Clusturing.csv')
X =dataset.iloc[:, [2,4]].values
```

```python
print(X)
```

## **Optimal Number of Clusters via DendoGrams**

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Age')
plt.ylabel('Euclidean distance')
```

## **Hierarchical Clustering Model Training on Trainig Set**

```python
from sklearn.cluster import  AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
yhc = hc.fit_predict(X)
```

```python
print(yhc)
```

## **Visualizing Clusters**

```python
plt.scatter(X[yhc == 0, 0], X[yhc == 0, 1], s = 100, c = 'magenta', label = 'Cluster 1')
plt.scatter(X[yhc == 1, 0], X[yhc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[yhc == 2, 0], X[yhc == 2, 1], s = 100, c = 'red', label = 'Cluster 3')
plt.scatter(X[yhc == 3, 0], X[yhc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[yhc == 4, 0], X[yhc == 4, 1], s = 100, c = 'yellow', label = 'Cluster 5')


plt.title('Cluster of Amazon users')
plt.xlabel('Age')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
